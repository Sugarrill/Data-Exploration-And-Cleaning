## Data Cleaning

```python
data.isnull().sum()
```

```python
data.isnull().sum() / len(data) * 100
```

```python
data.drop('Customer ID', axis=1,inplace = True)
```

```python
data.info()
```

```python
data['Description'].value_counts()
```

```python
data[data['Description'].isnull()]
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.duplicated().sum() / len(data) * 100
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data.info()
```

```python
data.loc[0, 'Price'] = data.loc[0, 'Price'].replace('$', '')
```

```python
data.loc[1, 'Price'] = data.loc[1, 'Price'].replace('$', '')
```

```python
data
```

```python
data['Price'] = data['Price'].astype('float64')
```

```python
data.info()
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Country'] = data['Country'].str.replace('UK', 'United Kingdom')
```

```python
data['Country'].value_counts()
```

```python
data.head()
```

```python
data.tail()
```

```python
data.drop('StockCode', axis=1,inplace=True)
```
